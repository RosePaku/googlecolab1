```python
# 원의 둘레와 넓이 구하기
pi=3.14159265 # 원주율
r=10 # 원의 반지름
print("원주율 =", pi)
print("반지름 =", r)
# 원의 둘레 = 2*pi*r
print("원의 둘레=", 2*pi*r)
# 원의 넓이 = pi*r*r
print("원의 둘레=", pi*r*r)
```

```python
# 복합대입연산자 (UPDATE)
no=100
no+=10 # 100 + 10 = 110 = no
no+=20 # 100 + 20 = 130 = no
no+=30 # 100 + 10 = 160 = no
print(no)
```

```python
# 복합대입연산자 => 문자
s="안녕"
s+="!" # 안녕! = s
s+="!" # 안녕!! = s
print(s)
```
