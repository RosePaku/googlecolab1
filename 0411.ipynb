```python
import numpy as np

a = input("구의 반지름을 입력해주세요 : ")
volume = (4/3) * np.pi * (float(a)**3)
print("구의 부피는 ", volume, "입니다") # 어떻게 해야 함?

surface_area = 4 * np.pi * (float(a) ** 2)  
sqrt_value = np.sqrt(surface_area)
print("구의 겉넓이는 {:.4f} 입니다".format(surface_area))

# np.sqrt() # 제곱근
```

```python
# 도전문제 1번 # 교수님
# 구의 부피와 겉넓이를 구하는 프로그램
pi = 3.14
r = float(input("구의 반지름을 입력해주세요: "))
# 부피
volume = (4/3) * pi * r**3
# 겉넓이
surface = 4 * pi * r**2
print("= 구의 부피는 {}입니다.".format(volume))
print ("= 구의 겉넓이는 {}입니다.".format(surface))
```

```python
# 계절을 알려주는 프로그램
import datetime
now = datetime.datetime.now()
# 봄
if 3 <= now.month <= 5:
   print("봄이야")
# 여름
if 6 <= now.month <= 8:
   print("여름이다")
# 가을
if 9 <= now.month <= 11:
   print("가을이다")
# 겨울
if 12 <= now.month <= 2:
   print("겨울왕국이다")
```

```python
# 계절을 알려주는 프로그램 // 상반기 하반기
import datetime
now = datetime.datetime.now()
# 봄
if 1 <= now.month <= 6:
   print("상반기야")
# 여름
if 7 <= now.month <= 12:
   print("하반기야")
```

```python
# 계절을 알려주는 프로그램 // 분기 (4분기)
import datetime
now = datetime.datetime.now()
# 봄
if 1 <= now.month <= 3:
   print("1분기야")
# 여름
if 4 <= now.month <= 6:
   print("2분기야")
# 가을
if 7 <= now.month <= 9:
   print("3분기야")
# 겨울
if 10 <= now.month <= 12:
   print("4분기야")
```
