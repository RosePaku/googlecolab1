```python
# 인덱스를 이용한 원소 선택
import pandas as pd
# 튜플 만들기
tuple_data = ("정대만","2023-04-01","남",True)
print(tuple_data)
```

```python
# 시리즈 만들기
sr=pd.Series(tuple_data, index=["이름","날짜","성별","상남자여부"])
print(sr)
```

```python
# 인덱스 이용
# 하나의 원소 선택
print(sr[3])
print(sr[0])
print(sr["상남자여부"])
print(sr["이름"])
```

```python
# 여러 개의 원소 선택
print(sr[[0,3]])
print(sr[[0,1,3]])
print("\t")
print(sr[["이름","상남자여부"]])
print("\t")
print(sr[["이름","날짜","상남자여부"]])
```

```python
# 여러 개의 원소 선택 with 범위 (슬라이싱)
print(sr[1:2+1])
print(sr[1:3])
print(sr[1:3])
print(sr["날짜":"성별"])
```

```python
# 행/열 삭제
import pandas as pd
# 데이터프레임 만들기
# 딕셔너리 => 데이터프레임
exam_data = {'국어':[100,50,80],
             '영어':[80,90,40],
             '수학':[60,40,50],
             '과학':[80,60,70]}
print(exam_data)
df = pd.DataFrame(exam_data,index=["서준","서진","서영"])
print(df)
```

```python
# 복사본
df1 = df[:]
# 행 삭제
df1.drop('서진',inplace = True)
# df1.drop('서진',axis=0)
print(df1)
```

```python
# 열 삭제
df2 = df[:]
df2.drop('과학',axis=1,inplace = True)
print(df2)
```

```python
df3 = df[:]
# 여러 개의 행 삭제
df3.drop(["서진","서진"],axis=0,inplace=True)
# 여러 개의 열 삭제
df3.drop(["국어","수학"],axis=1,inplace=True)
print(df3)
```

```python
# 같은 결과값 (복사본 만드는 방법2)
df3 = df.copy()
# 여러 개의 행 삭제
df3.drop(["서진","서진"],axis=0,inplace=True)
# 여러 개의 열 삭제
df3.drop(["국어","수학"],axis=1,inplace=True)
print(df3)
```

```python
# 행 선택
exam_data = {'국어':[100,50,80],
             '영어':[80,90,40],
             '수학':[60,40,50],
             '과학':[80,60,70]}
df=pd.DataFrame(exam_data,index=["서준","서진","서영"])
print(df)
# 행 하나를 선택
# loc => 인덱스 이름
name1 = df.loc['서영']
print(name1)
# iloc => 인덱스 숫자
no1 = df.iloc[2]
print(no1)

# 2개 이상의 행 선택
name2 = df.loc[["서준","서진"]]
print(name2)
no2 = df.iloc[[0,1]]
print(no2)

# 행 인덱스 범위 지정하여 선택
name3 = df.loc["서진":"서영"]
print(name3)
no3 = df.iloc[1:3]
print(no3)
```

```python
# 열 선택
print(df)
# 하나의 열 선택
df1 = df["영어"] # 시리즈
df2 = df[["영어"]] # 데이터프레임
print(df1);print(df2)
print(type(df1));print(type(df2))
# 여러 개의 열 선택
df3 = df[["국어","수학"]]
print(df3)
```

```python
# 원소 선택
print(df)
# 원소 1개 선택 = 행1 열1
# loc 이름
# 서영의 국어 점수
print(df.loc["서영",["국어","수학"]])
# iloc 번호
df.iloc[2,2:4]
```

```python
# 원소 2개 선택 = 행1 열2 with 범위 지정
# 서영이의 수학, 과학 점수
# loc 이름
print(df.loc["서영", "수학":"과학"])
# iloc 번호
print(df.iloc[2,2:4])
print(df.iloc[2,2:])
# 원소 여러 개 선택 = 행2 열2
# 서준, 서영이의 영어, 과학 점수
print(df.loc[["서준","서영"],["영어","과학"]])
print(df.iloc[[0,2],[1,3]])
# 범위 지정 : 서진, 서영이의 수학, 과학
print(df.loc["서진":"서영","수학":"과학"])
print(df.iloc[1:3,2:4])
```

```python
# 열 추가
print(df)
# 사회 과목 추가
df["사회"]=90
print(df)
# 체육 과목 추가
df["체육"] = [60,50,70]
print(df)
```

```python
# 행 추가
print(df)
# 같은 값을 가지고 있는 행 추가
df.loc["진우"] = 0
print(df)
# 여러 개의 값을 가지고 있는 행 추가
df.loc["민종"] = [90,90,90,100,80,90]
```

```python
# 원소 변경
print(df)
# 진우의 체육 점수를 변경
df.loc["진우","체육"] = 70
print(df)
# 진우의 사회 점수를 변경
df.iloc[3,4] = 80
print(df)
# 진우의 수학, 과학 점수를 변경
df.loc["진우",["수학","과학"]] = 80
print(df)
# 진우의 국어, 영어 점수를 변경
df.iloc[3,[0,1]] = 60, 70
print(df)
```

```python
# 행과 열 바꾸기 = 전치
print(df)
print(df.transpose)
print(df.T)
```

```python
# 인덱스 활용
# 특정 열을 행 인덱스로 설정
import pandas as pd
slamdunk = {"이름":["정대만","강백호","채치수"],
            "키":[183,189,193],
            "나이":[19,17,19],
            "평균득점":[25,5,15]}
df = pd.DataFrame(slamdunk)
print(df)
```

```python
# 행 인덱스 재배열
import pandas as pd 
dict_data = {"c0":[1,2,3],
             "c1":[4,5,6],
             "c2":[7,8,9],
             "c3":[10,11,12],
             "c4":[13,14,15]}
df=pd.DataFrame(dict_data, index = ["r0","r1","r2"])
print(df)
```

```python
# 행 인덱스 재배열
df1=df.reindex(["r0","r1","r2","r3","r4"])
print(df1)
df2=df.reindex(["r0","r1","r2","r3","r4"],fill_value = 0)
print(df2)
```

```python
# 행 인덱스 초기화
# 기존의 인덱스가 새로운 열로 추가
print(df)
df3 = df.reset_index()
print(df3)
```

```python
# 행 인덱스 정렬
print(df)
df4=df.sort_index(ascending=False) # 내림차순
print(df4)
```

```python
# 열 기준 정렬 = 열이 가지고 있는 값으로 정렬
print(df)
df5 = df.sort_values(by = "c1", ascending= False)
print(df5)
```
