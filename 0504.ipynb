## 시트활성화 후 열 확인

```python
# %%
# 시트 활성화 후 열 체크
ws = wb.active
# 열
ws["A"]
ws[2:3]
ws["A:C"]
```

## 반복문 활용해서 다중셀 접근

```python
# %%
# 반복문으로 다중셀 접근 
# 행 단위
for i in ws.iter_rows(min_row=2, max_row=3):
    print(i)
    for j in i:
      print(j)
```

```python
# %%
# 열 단위
for i in ws.iter_cols(min_row=1, max_row=2, max_col = 3):
    print(i)
    for j in i:
        print(j)

```

#엑셀 불러오기

```python
# %%
# 엑셀 불러오기
from openpyxl import load_workbook
wb = load_workbook(r"/content/drive/MyDrive/BDA/part6/stock price.xlsx", data_only=True)
ws = wb["Sheet1"]
```

```python
# %%
print(ws["A2"].value)
print(ws["D2"].value)
```

```python
# %%
# 셀 좌표로 값 출력
ws.cell(3,2).value
```

```python
# %%
ws_stock = ws["B2":"B11"]
for row in ws_stock:
    for cell in row:
        print(cell.value)

# 행이 열마다 투플 형태로 갖고 있기 때문에 이중 포문을 돌려야 함
```

```python
# %%
# 모든 행 값
for row in ws.rows:
    for cell in row:
        print(cell.value)
```

```python
# %%
# 모든 열 값
for col in ws.columns:
    for cell in col:
            print(cell.value)
```

```python
# %%
# 모든 행과 열 값 [DB 형태로 가져오기]
values = []
for row in ws.rows:
    row_values = []
    for cell in row:
        row_values.append(cell.value)
    values.append(row_values)
print(values)
```

```python
# %%
# 값 변경 후 저장
ws.cell(2,4,1000)
ws.cell(3,4,2000)
ws.cell(4,4,3000)

wb.save("stock_modify.xlsx")
```
